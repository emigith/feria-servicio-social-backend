```python
from PIL import Image

img = Image.open("assets/logo.jpg").convert("RGB")
img_small = img.resize((100, 100))

colors = img_small.getcolors(maxcolors=1000000)
colors = sorted(colors, reverse=True)

top_colors = colors[:10]

for count, color in top_colors:
    print(f"Cantidad: {count}, RGB: {color}, HEX: {'#%02x%02x%02x' % color}")
```
